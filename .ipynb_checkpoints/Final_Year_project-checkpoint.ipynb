```python
import pandas as pd
import numpy as np
from sklearn import ensemble, preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Load dataset 
train = pd.read_csv('train.csv')
Train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample = pd.read_csv('sampleSubmission.csv')
weather = pd.read_csv('weather.csv')
Weather = pd.read_csv('weather.csv')
spray = pd.read_csv("spray.csv")
Spray = pd.read_csv("spray.csv")
```

```python
Train.Date = pd.to_datetime(Train.Date)
Weather.Date = pd.to_datetime(Weather.Date)
Spray.Date = pd.to_datetime(Spray.Date)
```

```python
# Load map data
mdata = np.loadtxt("mapdata_copyright_openstreetmap_contributors.txt")
traps = pd.read_csv('Train.csv')[['Date', 'Trap','Longitude', 'Latitude', 'WnvPresent']]

aspect = mdata.shape[0] * 1.0 / mdata.shape[1]
lon_lat_box = (-88, -87.5, 41.6, 42.1)

# Plot map
plt.figure(figsize=(10,14))
plt.imshow(mdata, cmap=plt.get_cmap('gray'), extent=lon_lat_box, aspect=aspect)

# Spray locations
sprays = Spray[['Longitude', 'Latitude']].drop_duplicates()
sprays = sprays[sprays['Latitude'] < 42.3]  #outliers excluded
plt.scatter(sprays['Longitude'], sprays['Latitude'], marker='*', color='orange',alpha=.3, label='Spray Locations')

# Trap locations
T_loc = traps[['Longitude', 'Latitude']].drop_duplicates().values
plt.scatter(T_loc[:,0], T_loc[:,1], marker='x', label='Trap Locations')

# Weather locations
plt.scatter(x = (-87.933, -87.752), y = (41.995, 41.786), marker='o', color='r', label='Weather Station')
            
plt.title('West Nile Virus Preventions in Chicago')
plt.legend(frameon=1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.savefig('heatmap.png');
```

The blue Xs indicate traps placed for catching mosquitoes. Orange markings indicate areas that were sprayed and the two red circles are where the weather stations.

```python
# Get labels
labels = train.WnvPresent.values

# Not using codesum for this benchmark
weather = weather.drop('CodeSum', axis=1)

# Split station 1 and 2 and join horizontally
weather_stn1 = weather[weather['Station']==1]
weather_stn2 = weather[weather['Station']==2]
weather_stn1 = weather_stn1.drop('Station', axis=1)
weather_stn2 = weather_stn2.drop('Station', axis=1)
weather = weather_stn1.merge(weather_stn2, on='Date')

```

```python
# replace some missing values and T with -1
weather = weather.replace('M', -1)
weather = weather.replace('-', -1)
weather = weather.replace('T', -1)
weather = weather.replace(' T', -1)
weather = weather.replace('  T', -1)
```

```python
# Functions to extract month and day from dataset
# You can also use parse_dates of Pandas.
def create_month(x):
    return x.split('-')[1]

def create_day(x):
    return x.split('-')[2]

def create_year(x):
    return x.split('-')[0]
```

```python
train['month'] = train.Date.apply(create_month)
train['day'] = train.Date.apply(create_day)
test['month'] = test.Date.apply(create_month)
test['day'] = test.Date.apply(create_day)
train['year'] = train.Date.apply(create_year)
test['year'] = test.Date.apply(create_year)
```

```python
# Add integer latitude/longitude columns
train['Lat_int'] = train.Latitude.apply(int)
train['Long_int'] = train.Longitude.apply(int)
test['Lat_int'] = test.Latitude.apply(int)
test['Long_int'] = test.Longitude.apply(int)

```

```python
# drop address columns
train = train.drop(['Address', 'AddressNumberAndStreet','WnvPresent', 'NumMosquitos'], axis = 1)
test = test.drop(['Id', 'Address', 'AddressNumberAndStreet'], axis = 1)

```

```python
# Merge with weather data
train = train.merge(weather, on='Date')
test = test.merge(weather, on='Date')
train = train.drop(['Date'], axis = 1)
test = test.drop(['Date'], axis = 1)

```

```python
# Convert categorical data to numbers
lbl = preprocessing.LabelEncoder()
lbl.fit(list(train['Species'].values) + list(test['Species'].values))
train['Species'] = lbl.transform(train['Species'].values)
test['Species'] = lbl.transform(test['Species'].values)
```

```python
lbl.fit(list(train['Street'].values) + list(test['Street'].values))
train['Street'] = lbl.transform(train['Street'].values)
test['Street'] = lbl.transform(test['Street'].values)
```

```python
lbl.fit(list(train['Trap'].values) + list(test['Trap'].values))
train['Trap'] = lbl.transform(train['Trap'].values)
test['Trap'] = lbl.transform(test['Trap'].values)
```

```python
# drop columns with -1s
train = train.loc[:,(train != -1).any(axis=0)]
test = test.loc[:,(test != -1).any(axis=0)]
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
```

```python
# Random Forest Classifier 
clf = ensemble.RandomForestClassifier(n_jobs=-1, n_estimators=1000, min_samples_split=2)
clf.fit(train, labels)
predictions = clf.predict_proba(test)[:,1]
predictions
```

```python
#AdaBoost
ada = ensemble.AdaBoostClassifier(n_estimators=100)
ada.fit(train,labels)
predict1 = ada.predict_proba(test)[:,1]

predict1
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from matplotlib.pyplot import figure
```

```python
X=train
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X)
PCA_df = pd.DataFrame(data = X_pca, columns = ['PC1', 'PC2'])
PCA_df.insert(2,"WnvPresent",Train.WnvPresent.values)
PCA_df['WnvPresent'] = LabelEncoder().fit_transform(PCA_df['WnvPresent'])
PCA_df.head()
```

```python
figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')

classes = [1, 0]
colors = ['r', 'b']
for clas, color in zip(classes, colors):
    plt.scatter(PCA_df.loc[PCA_df['WnvPresent'] == clas, 'PC1'], 
                PCA_df.loc[PCA_df['WnvPresent'] == clas, 'PC2'], 
                c = color)
    
plt.xlabel('Principal Component 1', fontsize = 12)
plt.ylabel('Principal Component 2', fontsize = 12)
plt.title('2D PCA', fontsize = 15)
plt.legend(['NotPresent', 'Present'])
plt.grid()
```

```python
N = 30498
I = 13135
print(N)
print(I)
S = N - I
beta = 0.485
gamma = 0.31

sus = [] # infected compartment
inf = [] # susceptible compartment


def infections(S, I, N):
    t = 0
    while (t < 730):
        S = S - beta * ((S * I / N)) + (gamma * I)
        I = I + beta * ((S * I) / N) - (gamma * I)

        sus.append(S)
        inf.append(I)
       
        t = t + 1
    print(I/N)
    print(I)
    print(S)
```

```python
infections(S, I, N)
figure = plt.figure()
figure.canvas.set_window_title('SIS model')
figure.add_subplot(211)
inf_line, =plt.plot(inf, label='I(t)')

sus_line, = plt.plot(sus, label='S(t)')
plt.legend(handles=[inf_line, sus_line])
plt.xlabel('no of days')
plt.ylabel('popuation')

plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0)) # use scientific notation



plt.show()
plt.savefig('2008.png')
```
